```python
# batsman with most runs
bats_gr = ipl.groupby('batsman')
bats_gr['batsman_runs'].agg(
    total_runs=('sum')
).sort_values(by='total_runs', ascending=False).head(3)
```

```python
# find top 10 batsmen with max no of sixes
six.groupby('batsman').agg(
    no_of_sixes = ('batsman_runs', 'count')
).sort_values(by='no_of_sixes', ascending=False).head(10)
```

```python
# batsman with the most 4s and 6s in the last 5 overs
ipl['over'].max()
last_overs = ipl[(ipl['over'] > 15) & ((ipl['batsman_runs'] == 4) | (ipl['batsman_runs'] == 6))]
clutch = last_overs.groupby('batsman').agg(
    {'match_id': 'count'}
).sort_values(by='match_id', ascending=False)
clutch
```

```python
ipl.replace('Rising Pune Supergiant', 'Rising Pune Supergiants', inplace=True)
```

```python
virat_k = ipl.loc[ipl['batsman']=='V Kohli']
virat_k.groupby('bowling_team')['batsman_runs'].sum().reset_index().sort_values('batsman_runs', ascending=False)
```

```python
matches = ipl.groupby(['batsman', 'match_id'])['batsman_runs'].sum()
matches.loc['V Kohli'].max()
```

```python
def max_score(player):
  matches = ipl.groupby(['batsman', 'match_id'])['batsman_runs'].sum()
  return f'The maximum score of {player} is {matches.loc[player].max()}'
```

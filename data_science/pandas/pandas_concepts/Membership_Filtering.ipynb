 Note:
- "axis=0" denotes axis along the index i.e (from top to bottom / vertical)

###When axis=0 each column is passed to the function i.e func operates on each column.

- "axis=1" denotes axis along the columns
i.e (from left to right)

###When axis=1 each row is passed to the function i.e func operates on each row.

default value for axis is always 0.

Note:
any() , all() -:

any() -> returns true if at least one of the values in a row/column is True or equivalent i.e (non-empty, and non-zero)

all() ->  returns true only if all the values in a row/column are True or equivalent i.e (non-empty, and non-zero)

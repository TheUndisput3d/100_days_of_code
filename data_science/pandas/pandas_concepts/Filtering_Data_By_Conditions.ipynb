###Row Filter Mask
- Row Boolean Mask contains as many items as there are rows in the dataframe


```python
row_filter_mask = df['account length'] > 100
row_filter_mask
```

Column Filter Mask


- Column boolean mask contains as many items as there are columns in the dataframe


```python
col_filter_mask = df.columns.str.startswith('t')
col_filter_mask
```

The length of the filter should match the length of the number of rows in the data frame

```python
df.loc[row_filter_mask, :]
```

Applying more than one filters

```python
filter1 = (df['account length'] > 100)
filter2 = (df['international plan'] == "yes")
```

## Using more than one conditions or filters
- OR condition with "|"

```python
df.loc[filter1 | filter2, :]
```

AND condition with "&"

```python
df.loc[filter1 & filter2, :]
```

```python
df.loc[filter1 & filter2, col_filter_mask]
```

- Filtering using "where". Instead of dropping the non-eligible rows, it replaces the values with NANs.
- Displays NANs for those rows which don't satisfy the condition

- You can either drop these rows or replace NANs with some other value

```python
df.where(filter1 & filter2)
```

```python
df.where(filter1 & filter2).dropna()
```

```python
%%time
df.where(filter1 & filter2).fillna(0, axis=0)
```

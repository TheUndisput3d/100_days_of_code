```python
df = pd.read_csv("../datasets/Tips100.csv")
df.head()
```

assign()

```python
def tip_perc(df):
  print(df.head(2))
  return round(df['tip'] /df['size'], 2)
```

```python
#Way-1
df.assign(tip_per_person = tip_perc, tip_percentage = round(df['tip']/df['total_bill'], 2))
```

```python
#Way-2
df.assign(tip_per_person = lambda df: tip_perc(df), tip_percentage = round(df['tip']/df['total_bill'], 2))
```

```python
#Way-3
df = df.assign(tip_per_person = lambda df: round(df['tip'] /df['size'], 2), tip_percentage = round(df['tip']/df['total_bill'], 2))
```

```python
#simple use-case of kwargs, and passing funcs as parameters
def sum(**kwargs):
  sum = 0
  for key, val in kwargs.items():
    sum += val
  return sum

def mean(func, **kwargs):
  print(func(**kwargs) / len(kwargs))

mean(sum, **{'val1': 1, 'val2': 2, 'v': 3})
```

```python
nested_categories = [
    ["Alice", "Bob", "Charlie", "David"],     # People
    ["Apple", "Banana", "Cherry", "Date"],    # Fruits
    ["Cat", "Dog", "Elephant", "Fox"],        # Animals
    ["Rose", "Tulip", "Sunflower", "Cactus"]  # Plants
]
```

```python
for i in range(len(nested_categories)):
    print(nested_categories[i])
```

```python
[sorted(lst, key=len) for lst in nested_categories]
```

```python
[sorted(lst, key=lambda x : x[0]) for lst in nested_categories]
```

```python
lsts = []
for i in nested_categories:
    (lsts.append(sorted(i, key=lambda x : x[0])))
```

```python
indexed_values = [
    "index1", "index2", "index3", "index4", "index5"
]
list(map(lambda x: int(x[5:])**2, indexed_values))
```

```python
lst = list(map(lambda x: x**2, list(map(lambda x: int(x[5:]), indexed_values))))
```

```python
lst2 = list(filter(lambda x: x<10, lst))
lst2
```

```python
import functools as f
f.reduce(lambda a, b: a+b, lst2)/len(lst2)
```

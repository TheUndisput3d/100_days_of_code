## Map and Applymap

```python
df = pd.read_csv("/content/Tips100.csv")
df.head()
```

assign()

```python
def tip_perc(df):
  print(df.head(2))
  return round(df['tip'] /df['size'], 2)
```

```python
#Way-1
df.assign(tip_per_person = tip_perc, tip_percentage = round(df['tip']/df['total_bill'], 2))
```

```python
#Way-2
df.assign(tip_per_person = lambda df: tip_perc(df), tip_percentage = round(df['tip']/df['total_bill'], 2))
```

```python
#Way-3
df = df.assign(tip_per_person = lambda df: round(df['tip'] /df['size'], 2), tip_percentage = round(df['tip']/df['total_bill'], 2))
```

```python
#simple use-case of kwargs, and passing funcs as parameters
def sum(**kwargs):
  sum = 0
  for key, val in kwargs.items():
    sum += val
  return sum

def mean(func, **kwargs):
  print(func(**kwargs) / len(kwargs))

mean(sum, **{'val1': 1, 'val2': 2, 'v': 3})
```

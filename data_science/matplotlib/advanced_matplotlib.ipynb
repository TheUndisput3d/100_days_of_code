```python
# annotations example1
x = [x+1 for x in range(5)]
y = [a**2 for a in x]
size = [el for el in range(6,  12)]
color = ['pink', 'blue', 'black', 'orange', 'red']
plt.scatter(x, y)
plt.plot(x, y)
for i in range(len(x)):
  plt.text(x[i], y[i], 'aayush'+str(i), fontdict={'size': size[i], 'color': color[i]})
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
plt.xlabel('Average runs')
plt.ylabel('Strike rate')
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
plt.show()
```

```python
# Plotting 4 graphs using subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))


ax[0, 0].scatter(batters['avg'], batters['strike_rate'])
ax[0, 1].scatter(batters['avg'], batters['runs'])
ax[1, 0].hist(batters['avg'])
ax[1, 1].hist(batters['strike_rate'])


# ax[0].set_xlabel('Average Runs')
ax[0, 0].set_title('Avg_runs VS strike_rate')
ax[0, 0].set_ylabel('Strike Rate')
ax[0, 1].set_xlabel('average')


ax[0, 1].set_title('Avg_runs VS total_runs')
ax[0, 0].set_xlabel('average')
ax[0, 1].set_ylabel('Total Runs')

ax[1, 0].set_title('Average_histogram')
ax[1, 0].set_xlabel('average')
ax[1, 0].set_ylabel('Frequency')

ax[1, 1].set_title('Strike_rate_histogram')
ax[1, 1].set_xlabel('strike_rate')
ax[1, 1].set_ylabel('Frequency')

plt.tight_layout()

fig.show()
```

```python
# Create subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Scatter plot: avg vs strike_rate
ax[0, 0].scatter(batters['avg'], batters['strike_rate'], color='blue')
ax[0, 0].set_title('Avg vs Strike Rate')
ax[0, 0].set_xlabel('Average')
ax[0, 0].set_ylabel('Strike Rate')

# Scatter plot: avg vs runs
ax[0, 1].scatter(batters['avg'], batters['runs'], color='green')
ax[0, 1].set_title('Avg vs Runs')
ax[0, 1].set_xlabel('Average')
ax[0, 1].set_ylabel('Runs')

# Histogram: avg
ax[1, 0].hist(batters['avg'], bins=5, color='orange')
ax[1, 0].set_title('Histogram of Averages')
ax[1, 0].set_xlabel('Average')
ax[1, 0].set_ylabel('Frequency')

# Histogram: strike_rate
ax[1, 1].hist(batters['strike_rate'], bins=5, color='purple')
ax[1, 1].set_title('Histogram of Strike Rates')
ax[1, 1].set_xlabel('Strike Rate')
ax[1, 1].set_ylabel('Frequency')

# Adjust layout for better readability
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# another way to plot graphs, 3rd way

fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(2, 2, 1)
ax1.scatter(batters['avg'], batters['strike_rate'], color='red')

ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(batters['avg'], batters['runs'], color='green')

ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(batters['avg'], bins=5, color='purple')

ax4 = fig.add_subplot(2, 2, 4)
ax4.hist(batters['strike_rate'], bins=5, color='yellow')

fig.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data for plotting
x = np.random.rand(50) * 10  # Random x values
y = np.random.rand(50) * 10  # Random y values
z = np.random.rand(50) * 10  # Random z values

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
p = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# Labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')
fig.colorbar(p)

plt.tight_layout()
plt.show()
```

### 3D Surface Plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
```

```python
X.shape, Y.shape
```

```python
Z = X**2 + Y**2
Z.shape
```

```python
fig = plt.figure(figsize=(10, 6))
ax = plt.subplot(projection='3d')
g = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(g)
fig.show()
```

```python
Z = np.sin(X) + np.cos(Y)

fig = plt.figure(figsize=(10, 6))
ax = plt.subplot(projection='3d')
g = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(g)
fig.show()
```

```python
Z = np.sin(X)**2 + np.cos(Y)**2

fig = plt.figure(figsize=(10, 6))
ax = plt.subplot(projection='3d')
g = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(g)
fig.show()
```
